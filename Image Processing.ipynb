```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (replace with your dataset image path)
img = cv2.imread("E:/phd/lung cancer dataset/lung_colon_image_set/lung_image_sets/lung_n/lungn3.jpeg")

# Convert BGR (OpenCV default) → RGB (for matplotlib display)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Larger kernel size for more visible blur
gaussian_blur = cv2.GaussianBlur(img_rgb, (25, 25), 0)

fig, axs = plt.subplots(1, 2, figsize=(10,5))
axs[0].imshow(img_rgb)
axs[0].set_title("Original (Color)")
axs[0].axis("off")

axs[1].imshow(gaussian_blur)
axs[1].set_title("Gaussian Blur (25x25)")
axs[1].axis("off")

plt.show()


```

```python
#Sobel Edge Detection
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (replace with your dataset image path)
img = cv2.imread("E:/phd/lung cancer dataset/lung_colon_image_set/lung_image_sets/lung_n/lungn3.jpeg")

# Convert BGR (OpenCV default) → RGB (for matplotlib display)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobelx, sobely)
sobel_norm = cv2.convertScaleAbs(sobel_combined)

# Convert edges to 3-channel so we can overlay on color
sobel_color = cv2.cvtColor(sobel_norm, cv2.COLOR_GRAY2RGB)

plt.imshow(sobel_color)
plt.title("Sobel Edge Detection (Color Overlay)")
plt.axis("off")
plt.show()
```

```python
# Canny Edge Detection (side by side visualization)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (replace with your dataset image path)
img = cv2.imread("E:/phd/lung cancer dataset/lung_colon_image_set/lung_image_sets/lung_n/lungn3.jpeg")

# Convert BGR (OpenCV default) to RGB for displaying properly in matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale for processing
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Canny Edge Detection
canny_edges = cv2.Canny(gray, 100, 200)

# Plot Original and Canny next to each other
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(img_rgb)
axs[0].set_title("Original Image (Color)")
axs[0].axis("off")

axs[1].imshow(canny_edges, cmap="gray")
axs[1].set_title("Canny Edge Detection")
axs[1].axis("off")

plt.tight_layout()
plt.show()

```

```python
# Thresholding (Binary)
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (replace with your dataset image path)
img = cv2.imread("E:/phd/lung cancer dataset/lung_colon_image_set/lung_image_sets/lung_n/lungn45.jpeg")

# Convert BGR → RGB for matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale for processing
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Binary thresholding
_, binary_thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Show Original vs Binary side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(img_rgb)
axs[0].set_title("Original Image (Color)")
axs[0].axis("off")

axs[1].imshow(binary_thresh, cmap="gray")
axs[1].set_title("Binary Thresholding")
axs[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
# Otsu's Thresholding
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image (replace with your dataset image path)
img = cv2.imread("E:/phd/lung cancer dataset/lung_colon_image_set/lung_image_sets/lung_n/lungn45.jpeg")

# Convert BGR → RGB for matplotlib display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale for processing
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Otsu's Thresholding (automatic)
_, otsu_thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Show Original vs Otsu side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

axs[0].imshow(img_rgb)
axs[0].set_title("Original Image (Color)")
axs[0].axis("off")

axs[1].imshow(otsu_thresh, cmap="gray")
axs[1].set_title("Otsu's Thresholding")
axs[1].axis("off")

plt.tight_layout()
plt.show()
```
